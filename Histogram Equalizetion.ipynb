```python
import cv2
from skimage import io
from matplotlib import pyplot as plt
```

```python
img = cv2.imread("Q1_2.tif", 1)
```

```python
img1 = cv2.imread("Q1_3.tif",1)
```

```python
lab_img= cv2.cvtColor(img, cv2.COLOR_BGR2LAB)
```

```python
lab_img1= cv2.cvtColor(img1, cv2.COLOR_BGR2LAB)
```

```python
l, a, b = cv2.split(lab_img)
```

```python
l1, a1, b1 = cv2.split(lab_img1)
```

```python
equ = cv2.equalizeHist(l)
```

```python
equ1 = cv2.equalizeHist(l1)
```

```python
updated_lab_img1 = cv2.merge((equ,a,b))
```

```python
updated_lab_img3 = cv2.merge((equ1,a1,b1))
```

```python
hist_eq_img = cv2.cvtColor(updated_lab_img1, cv2.COLOR_LAB2BGR)
```

```python
hist_eq_img1 = cv2.cvtColor(updated_lab_img3, cv2.COLOR_LAB2BGR)
```

```python
clahe = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(8,8))
clahe_img = clahe.apply(l)
#plt.hist(clahe_img.flat, bins=100, range=(0,255))
```

```python
clahe1 = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(8,8))
clahe_img1 = clahe1.apply(l1)
```

```python
updated_lab_img2 = cv2.merge((clahe_img,a,b))
```

```python
updated_lab_img4 = cv2.merge((clahe_img1,a1,b1))
```

```python
CLAHE_img = cv2.cvtColor(updated_lab_img2, cv2.COLOR_LAB2BGR)
```

```python
CLAHE_img1 = cv2.cvtColor(updated_lab_img4, cv2.COLOR_LAB2BGR)
```

```python
cv2.imwrite('equalizedImage1.jpg', hist_eq_img)
cv2.imwrite('clhe1.jpg', CLAHE_img)
```

```python
plt.subplot(2,2,1),plt.imshow(img,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,2),plt.imshow(hist_eq_img,cmap = 'gray')
plt.title('Histogram Equalization'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,3),plt.imshow(CLAHE_img,cmap = 'gray')
plt.title('contrast limited histogram equalization'), plt.xticks([]), plt.yticks([])

plt.show()
```

```python
cv2.imwrite('equalizedImage2.jpg', hist_eq_img1)
cv2.imwrite('clhe2.jpg', CLAHE_img1)
```
