```python
import cv2
from skimage.filters import median
import numpy as np
from matplotlib import pyplot as plt
```

```python

img_salt_pepper_noise = cv2.imread('Q1_4.tif', 0)
```

```python
img = img_salt_pepper_noise
```

```python
median_using_cv2 = cv2.medianBlur(img, 9)
```

```python
cv2.imwrite('denoise_salt_&_papper.jpg', median_using_cv2)
```

```python
plt.subplot(2,2,1),plt.imshow(img,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,2),plt.imshow(median_using_cv2,cmap = 'gray')
plt.title('Denoised'), plt.xticks([]), plt.yticks([])

plt.show()
```
