```python
import cv2
from skimage import io, img_as_float
from skimage.filters import gaussian
import numpy as np
from matplotlib import pyplot as plt
```

```python
img1 = cv2.imread("Q1_1.tif")
```

```python
gaussian_using_cv2 = cv2.GaussianBlur(img1, (9,9), 0, borderType=cv2.BORDER_CONSTANT)
```

```python
cv2.imwrite('denoise.png', gaussian_using_cv2)
```

```python
plt.subplot(2,2,1),plt.imshow(img1,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,2),plt.imshow(gaussian_using_cv2,cmap = 'gray')
plt.title('Denoised'), plt.xticks([]), plt.yticks([])

plt.show()
```
